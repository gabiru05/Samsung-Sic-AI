```python
# Espacio disponible para desarrollar el código y agregar los comentarios necesarios.

def coin_change(coins, amount):
    ##esta sera la billetera donde iremos guardando las monedas necesarias para el cambio
    changes = []
    #empezamos por la moneda mas grande
    largest = 0
    while amount > 0:
        print("antes del if")
        print(amount)
        #si la moneda mas grande es mayor cambie a la siguiente
        if amount < coins[largest]:
            largest += 1
            print("priemer if")
            print(coins[largest])
            
        #Si la moneda es menor ira restando poco a poco devolviendo el cambio exacto
        #ademas se ira guardando en la billetera
        else:
            changes.append(coins[largest])
            amount -= coins[largest]
            print("segundo if")
            print(coins[largest])

    return changes

coins = [11, 15, 1]
amount = int(input("Input the amount: "))
changes = coin_change(coins, amount)
print(changes, len(changes))
```

```python
# Espacio para desarrollar el código y agregar los comentarios necesarios.

def find_minimum(row, col, triangle):
    
    #caso base cuando se llega aqui termina el codigo. se llego al final del triangulo
    #cuando baja a la siguiente columna se desea volver a evaluar todos la fila por eso devolvemos al inicio
    if row == len(triangle):
        print("0")
        return 0
        
    
    #caso recursivo se estara repitiendo hasta evaluar todo los posibles --menores numeros--
    else:
        #se evaluara entre derecha e izquierda mientras vaya descendiendo cual ofrece el --menor posible--
        #gracias a la funcion min, elegira entre la opcion y se quedara con esta.
        
        
        
        minimum = min(find_minimum(row + 1, col, triangle),
                    find_minimum(row + 1, col + 1, triangle))
        
        print(f"Row: {row}, Col: {col}")
        print(f"valor: {triangle[row][col]}")
        print("volvio\n")
        
        return triangle[row][col] + minimum
    

triangle = [
    [2],
    [3, 14],
    [6, 5, 7],
    [4, 20, 8, 3]
]

minimum = find_minimum(0, 0, triangle)
print("The minimum cost is ", minimum)
```

3. la funcion recursiva se encarga de recorrer todo los posibles sin tener que programar mucho
es cierto que requiere tener una buena logica y puede aumentar su dificultad pero suele ser lo mas optimo en muchos casos
usar funciones recursiva

```python
# Espacio para desarrollar el código y agregar los comentarios necesarios.

def bin_search(nums, x):

    low, high = 0, len(nums) - 1
    #empezamos dividiendo la lista en dos mitades y obteniendo el centro 
    #si el numero buscado se encuentra en la mitad termina
    #de lo contrario escogera en la mitad donde posiblemente este y descartara el recorre la otra.
    while low <= high:
        mid = (low + high) // 2
        if nums[mid] == x:
            return mid
        elif nums[mid] > x:
            high = mid - 1
        else:
            low = mid + 1
        
    return -1


nums=[1,2,3,3,5,6,7,8,9]
x = int(input("Ingresar objetivo del 1 al 9: "))

encontrado = bin_search(nums, x)
if encontrado == -1:
  print("No se encuentra el numero buscado.")
else:
  print(f"El numero {x} esta en la posicion {encontrado}")
```
