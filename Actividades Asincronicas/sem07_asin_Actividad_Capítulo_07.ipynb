# Actividad: Importación de módulos y su uso.


## Instrucciones

1. Importe los módulos que te permitan trabajar con fechas y tiempo (modulo datetime).

2. Use los diferentes métodos relacinados a operaciones con fechas y hora.

3. En el siguiente apartado revisa cada una las fechas contenidas en la lista e imprima el número de semanas y los dias restantes para llegar a esa fecha.

```python
from datetime import datetime

# Lista de fechas
lista_de_fechas = ['04/01/2024', '01/28/2024', '03/16/2024', '05/24/2024', '07/30/2025']

# Obtener la fecha actual
fecha_actual = datetime.now()

# Iterar sobre la lista de fechas
for fecha_str in lista_de_fechas:
    
    # Convertir la cadena de fecha en un objeto datetime
    fecha_obj = datetime.strptime(fecha_str, '%m/%d/%Y')
    
    # Calcular la diferencia de días entre la fecha actual y la fecha de la lista
    diferencia = fecha_obj - fecha_actual
    
    # Calcular el número de semanas y los días restantes
    semanas = diferencia.days // 7
    dias_restantes = diferencia.days % 7
    
    # Imprimir los resultados
    print(f"Para la fecha {fecha_str}: {semanas} semanas y {dias_restantes} días restantes.")


```

# Actividad: Manejo de la biblioteca Pandas.


## Instrucciones


1. Crear una serie de pandas, donde su indice este compuesto por fechas y en columna principal almace los valoras cargados en una lista.

2. Graficar los valores de la lista en función de la fecha.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Lista de precios
precios = [140, 130, 120, 115, 160, 110, 170, 150, 157, 135]

# Lista de fechas (pueden ser generadas de acuerdo a un rango de tiempo)
fechas = pd.date_range(start='2024-01-01', periods=len(precios), freq='M')

# Crear la serie de pandas
serie_precios = pd.Series(precios, index=fechas)

# Graficar los valores en función de la fecha
serie_precios.plot(marker='D', linestyle='--', color='yellow')
plt.title('Valores en función de la fecha')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.grid(True)
plt.show()
```

# Actividad: Operaciones con Data Frames


## Instrucciones


1. Dado el siguiente diccionario construya un objeto DataFrame.
2. Filtre los resultados por año de publicación.
3. Obtenga algunas columnas basado en que contenga alguna palabra en el nombre de la publicacion.


```python
import pandas as pd

# Diccionario de publicaciones literarias
publicaciones_literarias = {
    1605: "Don Quijote de la Mancha - Miguel de Cervantes",
    1865: "Alicia en el país de las maravillas - Lewis Carroll",
    1949: "1984 - George Orwell",
    1954: "El Señor de los Anillos - J.R.R. Tolkien",
    1997: "Harry Potter y la piedra filosofal - J.K. Rowling",
    2003: "El Código Da Vinci - Dan Brown"
}

# Construir DataFrame a partir del diccionario
df = pd.DataFrame(list(publicaciones_literarias.items()), columns=['Año', 'Publicación'])

# Filtrar los resultados por año de publicación
anno_publicacion = 1950
df_filtrado_por_anno = df[df['Año'] >= anno_publicacion]


df_filtrado_por_anno
```

```python
# Obtener algunas columnas basadas en una palabra clave en el nombre de la publicación
palabra_clave = 'Harry'
df_filtrado_por_palabra_clave = df[df['Publicación'].str.contains(palabra_clave)]

# Mostrar resultados
print("DataFrame original:")
print(df)
print("\nPublicaciones con año de publicación posterior a", anno_publicacion, ":")
print(df_filtrado_por_anno)
print("\nPublicaciones que contienen la palabra clave '", palabra_clave, "':")
print(df_filtrado_por_palabra_clave)
```
